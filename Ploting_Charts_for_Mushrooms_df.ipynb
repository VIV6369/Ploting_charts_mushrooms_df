***Exploring Mushrooms***

Mushrooms exist in a variety of different colors, shapes, sizes, textures, etc. In this project, you will analyze an extensive mushroom dataset from UCI using bar charts and acquaint yourself with the diverse array of mushrooms that exist worldwide.

Load mushroom_data.csv dataframe read through this table to get a sense of the type(s) of variables in the data and the structure of the table. 

```python
# Load libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#1 load in the data
df = pd.read_csv("mushroom_data.csv")
print(df.head())
```

There are 23 variables in this dataset (one for each column). Graphing each one individually would be tedious; luckily, you will use loops.

* Create an attribute called columns. This attribute returns the name of each variable in mushroom_data.csv.

* Create a loop that traverses each column in the columns list.

* Plot diffetent type of charts, clean it and make it easy readable. Get rid of non informative charts.

```python
columns = df.columns.tolist()

for i in columns:
  #Skips not informative charts
  if i == "Veil Type" or i == "Veil Color":
    continue
  # Creates pie charts for columns with less than 6 variables
  elif df[i].nunique()  < 6:
    plt.pie(df[i].value_counts(), labels = df[i].unique(), autopct = '%d%%', pctdistance=0.91)
    plt.axis('equal')
    plt.title(i)
# Creates bar charts for columns above 6 variables
  else:
    sns.countplot(df[i], order = df[i].value_counts(ascending=False).index)
    plt.title(i)
  # rotates the value labels slightly so they don’t overlap, also slightly increases font size
    plt.xticks(rotation=30, fontsize=10)
  # Hide x axis Label
    plt.xlabel('')
  plt.show()
  plt.clf()
```
